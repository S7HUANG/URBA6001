```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#2015-19&2020 contrast

def scatterplot_post(path):
    df2 = pd.read_csv(path) 
    max1 = int(df2.iloc[:,1:].max().max())
    min1 = int(df2.iloc[:,1:].min().min())           
    a = np.arange(min1,max1)
    fig, ax = plt.subplots(figsize=(7,7),dpi=200)
    ax.set_aspect('equal', adjustable='box')
    plt.grid(ls = '--',linewidth=0.8)
    plt.plot(a,a,ls = '--',c ='r',linewidth=1.5)

    ax.set_title(path.split('_')[1]+'(AQS, Ground sites)', fontsize=20, verticalalignment='bottom')
    ax.set_xlabel("2015-19(ug/m³)", fontsize=17)
    ax.set_ylabel("2020(ug/m³)", fontsize=17)

    x = df2['2015-19']
    y = df2['2020']
    for i in range(len(x)):
        if x[i]/y[i] <1:    
            plt.scatter(x[i],y[i],facecolors='none',s=60,edgecolors = 'r')
        if x[i]/y[i] >1:
            plt.scatter(x[i],y[i],facecolors='none',s=60,edgecolors = 'b')
    plt.show()

#scatterplot_post(r'ALL_NO2_cleanednew.csv')
#scatterplot_post(r'ALL_O3_cleanednew.csv')
#scatterplot_post(r'ALL_PM10_cleanednew.csv')
scatterplot_post(r'ALL_PM2.5_cleanednew.csv')
```
