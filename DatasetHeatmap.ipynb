```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('ALL_NO2.csv')
df.ReadingDateTime = pd.to_datetime(df.ReadingDateTime,format = ("%d/%m/%Y %H:%M"))
df[['year','month','day']] = df.ReadingDateTime.apply(lambda x: pd.Series(x.strftime("%Y,%m,%d").split(",")))
df1 = df[df['year']=='2021'] 
df2 = df1[df1['month']=='10']  
print(df2)
```

```python
NO2=df2.groupby('day')['Value'].mean()
```

```python
df3 = pd.read_csv('ALL_O3.csv')
df3.ReadingDateTime = pd.to_datetime(df3.ReadingDateTime,format = ("%d/%m/%Y %H:%M"))
df3[['year','month','day']] = df3.ReadingDateTime.apply(lambda x: pd.Series(x.strftime("%Y,%m,%d").split(",")))
df4 = df3[df3['year']=='2021'] 
df5 = df4[df4['month']=='10']  
print(df5)
```

```python
O3=df5.groupby('day')['Value'].mean()
```

```python
df6 = pd.read_csv('ALL_PM2.5.csv')
df6.ReadingDateTime = pd.to_datetime(df6.ReadingDateTime,format = ("%d/%m/%Y %H:%M"))
df6[['year','month','day']] = df6.ReadingDateTime.apply(lambda x: pd.Series(x.strftime("%Y,%m,%d").split(",")))
df7 = df6[df6['year']=='2021'] 
df8 = df7[df7['month']=='10']  
print(df8)
```

```python
PM2=df8.groupby('day')['Value'].mean()
```

```python
df9 = pd.read_csv('ALL_PM10.csv')
df9.ReadingDateTime = pd.to_datetime(df9.ReadingDateTime,format = ("%d/%m/%Y %H:%M"))
df9[['year','month','day']] = df9.ReadingDateTime.apply(lambda x: pd.Series(x.strftime("%Y,%m,%d").split(",")))
df10 = df9[df9['year']=='2021'] 
df11 = df10[df10['month']=='10']  
print(df11)
```

```python
PM10=df11.groupby('day')['Value'].mean()
```

```python
data1=pd.concat([NO2.rename('NO2'), O3.rename('O3'), PM2.rename('PM2.5'), PM10.rename('PM10')], axis=1)
```

```python
data1_T=data1.T
print(data1_T)
data1_T.to_csv('2021.10.csv')
```

```python
plt.ylabel = ["NO2", "O3", "PM2.5","PM10"]
fig, ax = plt.subplots(figsize=(7,7),dpi=200)
im = ax.imshow(data1.T,interpolation = 'nearest',cmap="Blues")
ax.set_yticks(np.arange(len(plt.ylabel)))
ax.set_yticklabels(plt.ylabel,fontsize=8, minor=False)
#plt.xlim(["NO2", "O3", "PM2.5","PM10"])
#plt.ylim([-20, 20])
#plt.yticks(fontsize=20)
```

```python
data = pd.read_csv('lodonmobility.csv')
data.date = pd.to_datetime(data.date)
data[['year','month','day']] = data.date.apply(lambda x: pd.Series(x.strftime("%Y,%m,%d").split(",")))
data2 = data[data['year']=='2021'] 
data3 = data2[data2['month']=='11'] 
data3=data3[['day','driving','transit','walking']] 
print(data3)
```

```python
data4 = pd.read_excel('Add2.xlsx')
data5=pd.concat([data3, data4], axis=0)
mobility=data5[['driving','transit','walking']]
mobility_T=mobility.T
print(mobility_T)
```

```python
mobility_T.to_csv('mobility2021.11.csv')
```

```python
plt.ylabel = ['driving','transit','walking']
fig, ax = plt.subplots(figsize=(7,7),dpi=200)
im = ax.imshow(mobility.T,interpolation = 'nearest',cmap="Blues")
#ax.set_yticklabels(plt.ylabel,fontsize=8, minor=False)
ax.set_yticks(np.arange(len(plt.ylabel)))
ax.set_yticklabels(plt.ylabel,fontsize=8, minor=False)
```
